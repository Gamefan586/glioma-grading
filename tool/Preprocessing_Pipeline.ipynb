```python
import pandas as pd
```

```python
df_gbm = pd.read_csv("/Users/ronnitrana/ACSEF/glioma-grading/data/Firebrowse/Glioblastoma Multiforme/mRNAseq Analyses/RSEM_genes_transposed_raw_count_Hybridization_REF_deleted.csv")
df_lgg = pd.read_csv("/Users/ronnitrana/ACSEF/glioma-grading/data/Firebrowse/Lower-Grade Glioma/mRNAseq Analyses/RSEM_genes_transposed_raw_count_Hybridization_REF_deleted.csv")
```

```python
df_gbm
```

```python
df_lgg
```

```python
import numpy as np
combined_avg = np.array([df_lgg.mean(), df_gbm.mean()])
```

```python
threshold = pd.DataFrame(combined_avg, columns = df_lgg.columns.tolist())
```

```python
threshold = threshold.mean()
```

```python
transformed_lgg = pd.DataFrame(np.where(df_lgg > threshold, 1 , 0 ) , columns = df_lgg.columns.to_list())
```

```python
transformed_gbm = pd.DataFrame(np.where(df_gbm > threshold, 1 , 0 ) , columns = df_lgg.columns.to_list())
```

```python
transformed_lgg
```

```python
transformed_gbm
```

```python
transformed_gbm["classification"] = pd.Series(1, index=transformed_gbm.index)
```

```python
transformed_lgg["classification"] = pd.Series(0, index=transformed_lgg.index)
```

```python
combined_lgg_gbm = pd.concat([transformed_gbm, transformed_lgg])
```

```python
combined_lgg_gbm.to_csv("/Users/ronnitrana/ACSEF/glioma-grading/data/Firebrowse/RSEM_genes_transposed_raw_count_Hybridization_REF_deleted.csv"
```
